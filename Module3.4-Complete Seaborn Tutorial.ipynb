##### Data Visualization Library

##### Importing libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built in dataset

```python
# Total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips.head()
```

```python
# visualize the data
sns.relplot(data = tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
# Setting a theme for your plots
sns.set_theme()
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

#### Scatter Plot

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length', hue='species', data=iris)
```

```python
#Loading the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
# Count Plot
sns.countplot(x = 'class', data=titanic)
```

#### Bar Chart

```python
# house price dataset
from sklearn.datasets import fetch_california_housing
house_cali =fetch_california_housing()

house = pd.DataFrame(house_cali.data, columns = house_cali.feature_names)
house['PRICE'] = house_cali.target 
```

```python
house.head()
```

```python
house.shape
```

```python
print(house_cali)
```

#### Distribution Plot

```python
sns.distplot(house['PRICE'])
```

#### Corelation Matrix
- Positive Corelation
- Negative Corelation

```python
# AveRooms , HouseAge

correlation = house.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot = True, annot_kws={'size': 8}, cmap='Blues')
```
