#### Creating Pandas DataFrame 

```python
# importing the boston house price data 
import pandas as pd
import numpy as np

# Load Boston Housing data
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

# Feature names (13 columns)
column_names = [
    "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
    "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"
]

# Create DataFrame
boston_df = pd.DataFrame(data, columns=column_names)
boston_df['MEDV'] = target  # MEDV is the target (Median value of owner-occupied homes)

# Show head
print(boston_df.head())
```

#### Correlation:
##### 1. Positive Correlation
If the correlation between two feattures is positive this means that if one featuere grows, the other will be proportionally grow as well. Example Number of Bedrooms and the House Price. More the Bedrooms --> More the price.

##### 2. Negative Correlation
If the correlation between two feattures is negative this means that if one featuere grows, the other will be  decrease. Example Number of Crimes and the House Price. More the crime --> Less the price
