## Matplotlib:

- Useful for making Plots

```python
import matplotlib.pyplot as plt
```

```python
# importing numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

#### Plotting the data

```python
# Sine wave
plt.plot(x,y)
plt.show()
```

```python
# Cosine Wave
plt.plot(x,z)
plt.show()
```

```python
# Adding title x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine value')
plt.title('Sine Wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('Cosine value')
plt.title('Cosine Wave')
plt.show()
```

```python
# Parabola 
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# plot with different symbols
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x,y, 'g.')
```

```python
plt.plot(x,y, 'rx')
```

```python
# Plot multiple curves in one graph
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

##### BarPlot


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['english', 'french', 'spanish', 'latin', 'german']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('Number of People')
plt.show()
```

##### PieChart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

##### Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color='g')
ax.scatter(x,z,color='b')
plt.show()
```

#### 3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap='Blues')
plt.show()
```
