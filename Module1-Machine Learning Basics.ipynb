#### What is Artificial Intelligence?

**Artificial Intelligence**: is a Branch of Computer Science that is concerned with building smart & intelligent Machines.

- Example of non intelligent machines: Watch, Bike (Machines cannot think on their own)
- Example of intelligent Machines: Autonomous cars, Virtual Assistants like Siri, Alexa etc. (Can make decisions)

#### What is Machine Learning?

**Machine Learning**: Machine Learning is a technique to implement AI that can learnn from the data by themselves without being explicitly programmed. Machine Learning is all about data. 
Example: we want a system to detect whether the images given to the machine is Cat or Dog. For the machine to detect if the image given to it is Cat or dog, it needs to learn just like humans. To make the Machine learn we make a Model (Program) that takes in hundreds of images of Cats and dogs and trains the machine how a cat and a dog looks like (by pattern detection). And once the model is trained now the machine can predict whether the new image is Cat or Dog (this is called inferencing).

#### What is Deep Learning?

**Deep Learning**: Deep learning is a subfield of Machine Leanring that uses Artifcial neural networks to learn from the data.

These Artificial Neural Networks are designed to mimic human brains. Like Human brains these ANN also contains Neurons that are internconnected. Each neuron process the information from the data and sends its output to another connected neurons. These are nothing but mathematical models. There Different layers in Artificial Neural Network, each layer is made up of multiple neurons. First layer is input layer where it recieves data as input. The other layers corrent to it are hidden layers (Hidden because we don't know what neuron will work after the first input layer) until the last layer which is output layer that produces output.


## Module 1.2 Types of Machine Learning 

#### Machine Learning?
**Machine Learning**: Machine Learning is a technique to implement AI that can learnn from the data by themselves without being explicitly programmed.

#### Types of Machine Learning
There are 3 main types of Machine Learning:
1. Supervised Learning
2. Unsupervised Learning
3. Reinforcement Learning



##### 1. Supervised Learning
In supervised learning, the Machine learning algorithm learns from **Labelled Data**.



##### 2. Unsupervised Learning
In Unsupervised learning, the Machine learning algorithm learns from **Unlabelled Data**.


##### 3. Reinforcement Learning
Reinforcement Learning is an area of Machine learning concerned with how intelligent agents take actions in an environment to maximize its rewards.
There are 4 main aspects in reinforcement learning:
- a) Environment 
- b) Agent
- c) Action
- d) Reward

The goal is to build an agent that acts in that environments. The agent will then try to take some action in that environment and for that action it gains some rewards. For example we want make a Software that can play chase like a human being. Here our chess board becomes an environment and out computer becomes the agent, which tries to take an action in this case moves on the chess board. For each step it will get rewarded depending on how good the move is. If the move is most profiting to the agent it will get the most reward if it is most profiting to the opponent the agent will recieve the least reward.


## Module 1.3 Supervised Learning

In Supervised learning, the Machine Learning Algorithm learns from **Labelled Data**.

**Labelled Data**: Labelled Data is the data that is already been classified or named after the label. The algorithm tries to map the label and the data so that it can recognize it.

##### **Types of Supervised Learning**:
There are two main types of Supervised learning:
1) Classification
2) Regression

###### 1) Classification:
Classification is about predicting a class or discrete values. for example: Male or Female; True or False: Dog or Cat etc.

###### 2) Regression:
Regression is about predicting a quantity or continous values. for example: Salary; Age; Price etc.



##### Some Important Supervised Learning Algorithms:
###### 1) Classification:
- Decision Tree Classification
- Random Forest Classification
- K-Nearest Neighbor

###### 2) Regression:
- Logistic Regression
- Polynomial Regression
- Support Vector Machines

## Module 1.4 Unsupervised Learning

In Unsupervised Learning, the Machine Learning Algortihm learns from **Unlabelled Data**.
Unlike Supervised Learning where we give supervision to the model, in unsupervised learning the model learns on its own and detects patterns from the data. 

##### **Types of Unsupervised Learning**:
There are two main types of unsupervised learning:
1) Clustering
2) Association

###### 1) Clustering:
Clustering is an unsupervised task which involves grouping the similar data points.

###### 2) Association:
Association is an unsupervised task that is used to find important relationships between data points.





##### Some Important Supervised Learning Algorithms:
- K-Means Clustering
- Hierarchial Clustering
- Principal Component Analysis (PCA) (Used to reduce dimensions of our data)
- Apriori
- Eclat

## Module 1.5 What is Deep Learning?

In this Module we will learn about Basics of deep learning, What is Neural network and what are the applications of Deep learning? 

#### What is Deep Learning ?

**Deep learning**: Deep learning is a subfield of Machine learning that uses Artificial neural networks to learn from the data.


The neural networks are made up of multiple layers. Each layer can have multiple neurons aligned. Each neuron has a mathematical function assigned to itself to do a particular task (to process the information). The neural network will have one input layer, one output layer and multiple hidden layers depending on the complexity of the task.

#### Machine Learning  vs Deep Learning

The main differnce between Machine learning and deep learning is **Feature extraction**.

**Feature Extraction**: Lets say we want machine learning model to predict whether an image represents a car.
When we feed this image to a machine learning model we need to tell the model that this features are important for it to be a car (eg. 4 wheels, shape, headlamps). In case of Deep learning we donot need to specify them manually the neural network can identify the features bu themselves using the mathematical function assigned to it.

**Application of Deep Learning**:

- Healthcare
- Autonomous cars
- Computer Vision
- Natural Language Processing
